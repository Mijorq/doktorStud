```python
import numpy as np
from matplotlib import pyplot as plt
```

# `matplotlib` pojęcia

O [matplotlib:](https://matplotlib.org/3.2.1/tutorials/introductory/usage.html#parts-of-a-figure)  
`https://matplotlib.org/3.2.1/tutorials/introductory/usage.html#parts-of-a-figure`

1. **Figure**: cały obszar zawierający w sobie wszytskie elementy rysunku, jak np. tytuł, legenda, płótno). Wewnątrz *figure* znajdują się elementy zwane *axes*. Obszar *figure* może zawierać jeden lub więcej elementów *axes*.
2. **Axes**: właściwy obszar rysunku, prezentujący dane. Obszar ten może mieć 2 lub 3 osie **Axis**.
3. **Axis**: to osie wykresu - 2 osie dla wykresu 2D i 3 dla 3D. Te elemnty są odpowiedzialne za ustawienia skali osi, ustawienia punktów na osiach i opisy osi.

> **Axes** i **Axis** to są różne pojęcia!!!

Owołania do **figure** i **axes**:
  - przy wielu prostych wykresach odbywa się automatycznie bez udziału użytkownika
  - przy bardziej zaawansowanym formatowaniu wykresu konieczne jest odwołanie jawne
  
### Odwołanie jawne

```python
# jawne wywołanie 'figure'
f = plt.figure()
plt.title('Okno "Figure" - rozmiar domyślny')

f = plt.figure(figsize=(15,4))
plt.title('Okno "Figure" - rozmiar 15 na 4 w calach')
print(f'Typ obiektu: {type(f)}')
```

```python
# jawne wywołanie 'figure' i 'axes'
f,ax = plt.subplots(2,2)

# spr typu 'axes' - tablica numpy
print(f'Typ obiektu ax:\t{type(ax)}\nWymiar:\t\t{ax.shape}')

# generuje tytuły typu 'ax_1',...
tyt = [f'ax_{i}' for i in range(1,5)]
print(f'Tytuły:\t\t{tyt}')

# dodanie tytułu do każdego 'axes'
for i,tytul in enumerate(tyt):
    ax.ravel()[i].set_title(tytul)
    
# automatyczne dopasowanie 'sub' wykresów do obszaru 'figure'
plt.tight_layout()
```

```python
# typ pojedynczego 'axes'
print(f'Typ : {type(ax[0,0])}')
```

### Odwołanie automatyczne

```python
x = np.arange(-3,3.1,0.5)
y = x**2
plt.plot(x,y);
```

# Podstawowe formatowanie wykresu

### Formatowanie [linii](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle)
  > - kolor linii: `color` lub `c`, np. c ='red' lub c = 'r'
  > - grubość linii: `linewidth` lub `lw`, np.  lw = 2
  > - styl linii: `linestyle` lub `ls`, np. ls = '--' (przerywana)

```python
plt.plot(x,y,c='red', ls = '-.', lw=3);
```

### Proste formatowanie osi

```python
plt.plot(x,y,c='red', ls = '-.', lw=3);

# ustawienie zakresu wartości na osiach
plt.xlim(-5,10)
plt.ylim(-2,10)
```

```python
plt.plot(x,y,c='red', ls = '-.', lw=3);

# ustawia etykiety osi
plt.xlabel('oś oX')
plt.ylabel('oś oY')

# ustawia etykiety punktów osi:
#  - tablica punktów -3,-2,..., lista etykiet ['a',...]
plt.xticks(np.arange(-3,3.1),['a','b','c','d','e','f','g']);
```

```python
plt.plot(x,y,c='red', ls = '-.', lw=3);

# zmiana skali wybranej osi np. y:
# "linear", "log", "symlog", "logit"
plt.yscale('log')
```

### Opisy

  - `plt.title()`: dodaje tytył do wykresu
  - `labela` i `plt.legend()`: dodaje legendę automatycznie pozycjonowaną

Równaia matematyczne można pisać:
  - używając składni `Latex-a`
  - umieszczjąc równanie wewnątrz znaków `$`
  - umieszczając cały zapis równania w składni Pythona jako surowy ciąg znaków, czyli: `r'surowy cąg znaków'`

Ten sposób [isania równań jest praktycznie taki sam jak w `markdown` np.:
  > - surowe równanie: `$y = x^2 \cdot \frac{1}{x+1}$`
  > - interpretowane równanie: $y = x^2 \cdot \frac{1}{x+1}$

```python
# rysuje 2 funkcje
plt.plot(x,y,label=r'Równanie $y=x^2\cdot \frac{a-c}{b^3}$',\
         c='red', ls = '-.', lw=3);

plt.plot(x,x**3/(x+10),label=r'Równanie: $\frac{x^3}{x+10}$',c='g',ls='--')

# tytuł wykresu
plt.title(r'Wykresy - np. $F(x) = \sum_{i=1}^nx^3$');

# dodawanie legendy
plt.legend()
```

```python
# ustawienie ziarna
np.random.seed(222)
# dane losowe z rozkładu normalnego: śr = 10, sigma = 3
y = np.round(np.random.normal(10,3,50),3)
x = np.arange(y.size)
print(f'x:\n{x}\n\ny:\n{y}')
```

```python
expr = ['plt.plot(x,y,"*r")','plt.scatter(x,y,c="g")','plt.bar(x,y)']
f = plt.figure(figsize=(15,4))
for i,ex in enumerate(expr,1):
    plt.subplot(1,3,i)
    exec(ex)
    plt.title(ex)
plt.tight_layout()
```
